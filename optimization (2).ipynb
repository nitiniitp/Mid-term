```python
import numpy as np
import matplotlib.pyplot as plt

# Cost function
def cost(r):
    return 6/r + 0.06*np.pi*r**2

# Radius values
r = np.linspace(0.5, 6, 500)
C = cost(r)

# Optimal radius
r_opt = (50/np.pi)**(1/3)
C_opt = cost(r_opt)

# Plot
plt.figure(figsize=(8,5))
plt.plot(r, C, label="Cost function C(r)")
plt.scatter(r_opt, C_opt, label="Minimum point", s=60)

plt.xlabel("Radius r (in)")
plt.ylabel("Cost ($)")
plt.title("Cost vs Radius for Cylinder Design")
plt.legend()
plt.grid(True)

plt.show()

print("Optimal radius:", r_opt)
print("Optimal height:", 60/(np.pi*r_opt**2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ranges
r = np.linspace(0.5, 6, 80)
h = np.linspace(0.5, 10, 80)

R, H = np.meshgrid(r, h)

# cost function
C = 0.1*np.pi*R*H + 0.06*np.pi*R**2

# 3D plot
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(R, H, C, cmap='viridis')

ax.set_xlabel("Radius r")
ax.set_ylabel("Height h")
ax.set_zlabel("Cost")
ax.set_title("3D Cost Surface")

plt.show()
```
